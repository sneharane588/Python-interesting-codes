```python
from fbchat import Client
from fbchat.models import Message
```

```python
username = "sneharane588@gmail.com"
password = "madhmashi"
client = Client(username, password) #client object
```

```python
#meth of Client obj
dir()
```

```python
# get a list of threads, a thread can be a user or a group
users = client.fetchThreadList()
print(users)
```

```python
#get all the information about these users
detailed_users = [list(client.fetchThreadInfo(user.uid).values()) [0]for user in users]
detailed_users
```

```python
#thread object have a message_count attribute to count no. of msgs betwn u & thread
# sort by number of messages
sorted_detailed_users = sorted(detailed_users, key = lambda u: u.message_count, reverse = True )
sorted_detailed_users
```

```python
best_frnd = sorted_detailed_users[0]
print("Best friend is", best_frnd.name,"with a msg count of",best_frnd.message_count)
```

```python
#send msg to best_frnd
client.send(Message(text=f"Congratulations {best_frnd.name}, you are my best friend with {best_frnd.message_count} messages!"),
            thread_id=best_frnd.uid)
```

```python
# to get all the users you talked to in messenger
all_users = client.fetchAllUsers()
print("You talked with a total of", len(all_users), "users!")
```
