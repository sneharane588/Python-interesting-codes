### _creating histogram_

```python
import numpy as np
import seaborn as sns

#generate normally distributed data

data = np.random.randn(1000)
sns.distplot(data, kde=True, rug=True)
```

```python
sns.set_style('dark')
sns.distplot(data, kde=True, rug=True) #using previous data
```

```python
import matplotlib.pyplot as plt
sns.distplot(data, kde=True, rug=True)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
#matplpot lib commands can be applieed to seaborn plots
```
