# <center>Data visualization using matplotlib</center><hr>

### _Line Plotting_

```python
from matplotlib import pyplot as plt

#x-axis values [x1,x2,x3,x4]
x = [5, 2, 9, 4, 7]

#y-axis values [y1,y2,y3,y4]
y = [10, 5, 8, 4, 2]

#function to plot
plt.plot(x,y)
plt.show()
```

### _bar plotting_

```python
plt.bar(x,y)
plt.show()
```

### _histogram plotting_

```python
#y-axis values [y1,y2,y3,y4]
y = [13, 1, 2, 6, 4]

plt.hist(y)
plt.show()
```

### _scatter plotting_

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')

#ro to make red circles
plt.axis([0, 6, 0, 20])
plt.show()
```
